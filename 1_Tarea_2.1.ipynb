```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python

Cambridge_data = pd.read_csv("Cambridge.csv")
pd.options.display.max_columns = None
Cambridge_data
```

```python
#Los registros de 5 "host name" diferentes
Hostname= Cambridge_data[Cambridge_data["host_name"].isin(["Janet", "Kevin", "Nicole", "Jimmy", "Greg"])]
Hostname

Hostname.to_csv("Ejercicio1_5Hostname.csv")
```

```python
#Los registros de los host que se hallan unido a Airbnb "despues del aÃ±o 2020"
Fecha=Cambridge_data[(Cambridge_data["host_since"] >= "2020-01-01")]
Fecha

Fecha.to_csv("Ejercicio2_Hostname2020.csv")
```

```python
#Los registros de los host que responden "a mÃ¡s tardar en 1 dÃ­a"
undia=Cambridge_data[(Cambridge_data["host_response_time"] == "within a day") | (Cambridge_data["host_response_time"] == "within an hour") | (Cambridge_data["host_response_time"] == "within a few hours")]
undia

undia.to_csv("Ejercicio3_Hostname1dia.csv")
```

```python
#Los registros de las filas MÃºltiplos de 200..(400, 600, 800, etc.)
multiplos200=Cambridge_data.iloc[range(200, len(Cambridge_data), 200)]
multiplos200

multiplos200.to_csv("Ejercicio4_Multiplos200.csv")
```

```python
#Los registros de los tipos de cuarto: "Entire home" y "Private room"
tiposcuarto=Cambridge_data[(Cambridge_data["room_type"] == "Entire home/apt") | (Cambridge_data["room_type"] == "Private room")]
tiposcuarto

tiposcuarto.to_csv("Ejercicio5_Tiposcuarto.csv")
```

```python
#Los registros que "a lo mucho cuenten con 4 camas"
camas4=Cambridge_data[(Cambridge_data["beds"] == 4.0)]
camas4

camas4.to_csv("Ejercicio6_Camas4.csv")
```

```python
#Los registros que "sean superhost"
superhost=Cambridge_data[(Cambridge_data["host_is_superhost"] == "t")]
superhost

superhost.to_csv("Ejercicio7_Superhost.csv")
```

```python
#Los registros de las columnas "impares"
columnasimpares=Cambridge_data.iloc[:, 1::2]
columnasimpares

columnasimpares.to_csv("Ejercicio8_Columnasimpares.csv")
```

```python
#Los regios de los host cuyo "host_responde_rate" sea mayor a 95% y que no sean "superhost"
Superhost=Cambridge_data[(Cambridge_data["host_response_rate"] >= 0.95) & (Cambridge_data["host_is_superhost"] == "f")]
Superhost

Superhost.to_csv("Ejercicio9_hostresponse_Superhost.csv")
```

```python
#Eliminar un signo de una columna 
Cambridge_data['host_response_rate']=Cambridge_data['host_response_rate'].str.replace('%', '')
```

```python
#ConversiÃ³n de tipo de dato de columna de tipo Object a Float
Cambridge_data['host_response_rate']= Cambridge_data['host_response_rate'].astype(float)
```

```python
#Los registros cuyas propiedades no tengan "1 baÃ±o"
unbaÃ±o=Cambridge_data[(Cambridge_data["bathrooms"] != 1)]
unbaÃ±o 

unbaÃ±o.to_csv("Ejercicio10_1BaÃ±o.csv")
```

```python
#Los registros que tengan un precio entre "$100 y $200 US"
precio=Cambridge_data[(Cambridge_data["price"] >= "$100") & (Cambridge_data["price"] <= "$200")]
precio

precio.to_csv("Ejercicio11_Precio100y200.csv")
```

```python
#Los registros que sean "instant bookable" y que hallan verificado su perfil "host identity _verified"
instant_identity=Cambridge_data[(Cambridge_data["instant_bookable"] == "t") & (Cambridge_data["host_identity_verified"] == "t")]
instant_identity

instant_identity.to_csv("Ejercicio12_InstantIdentity.csv")
```

```python
#Los registros de los tipos de cuarto: "hotel room" o "Private room"
hotel_private=Cambridge_data[(Cambridge_data["room_type"] == "Hotel room") | (Cambridge_data["room_type"] == "Private room")]
hotel_private

hotel_private.to_csv("Ejercicio13_HotelPrivate.csv")
```

```python
#Los registros que tengan mas de 4 "accommodates"
accommodates = Cambridge_data[(Cambridge_data["accommodates"] > 4)]
accommodates

accommodates.to_csv("Ejercicio14_Acommodates.csv")
```
